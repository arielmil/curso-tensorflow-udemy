```python
import tensorflow as tf
```

```python
vetor = tf.constant([5, 10, 15], name = 'vetor')
type(vetor)
```

```python
print(vetor)
```

```python
#Soma ao vetor [5, 10, 15], o vetor [5, 5, 5] ficando [10, 15, 20]
soma = tf.Variable(vetor + 5, name = 'soma')
```

```python
print(soma)
```

```python
valor = tf.Variable(0, name = 'valor')
```

```python
#Loop para somar 1 ao valor. Com eager execution, o valor é impresso a cada iteração
#Ou seja, não é mais necessário executar sessão para ver o resultado!
for i in range(5):
    valor = valor + 1
    print(valor.numpy())
```
