```python
import tensorflow as tf
```

Obs: Multiplcação faz a operação da seguinte forma:
a = [a1, a2, a3, ... , an], b = [b1, b2, b3, ... , bn]

tf.multiply(a, b) faz:

[a1 * b1, a2 * b2, a3 * b3 , ... , an * bn].

```python
a = tf.constant([[-1.0, 7.0, 5.0]], name = 'entradas')
b = tf.constant([[0.8, 0.1, 0.0]], name = 'pesos')

multiplicacao = tf.multiply(a, b)
soma = tf.reduce_sum(multiplicacao)
```

Tentar fazer a multiplicação entre inteiros e floats retorna um erro:

```python
a = tf.constant([[1.0, 2.0, 3.0]], name = 'entradas')
b = tf.constant([[4, 5, 6]], name = 'pesos')

try:
    mult = tf.multiply(a, b)
except Exception as e:
    print(e)
```

```python
b = tf.constant([[4.0, 5.0, 6.0]], name = 'pesos')

mult = tf.multiply(a, b)
print(mult)
```

Tentar fazer multiplicação de objetos de tamanhos incompativeis retorna um erro:

```python
b = tf.constant([[4.0, 5.0, 6.0, 7.0]], name = 'pesos')

try:
    mult = tf.multiply(a, b)
except Exception as e:
    print(e)
```

tf.reduce_sum faz o produto escalar entre dois vetores.
Ou seja:

sum = [a1, a2, a3, ... , an] = a1 + a2 + a3 + ... + an

Nesse caso, sum = 4 + 5 + 6 = 15.

```python
b = tf.constant([[4.0, 5.0, 6.0]], name = 'pesos')
sum = tf.reduce_sum(b)
sum
```
