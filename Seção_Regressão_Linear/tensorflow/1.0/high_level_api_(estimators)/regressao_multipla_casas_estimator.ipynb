```python
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
base = pd.read_csv('house_prices.csv', usecols = colunas_usadas)
```

```python
base.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
X = base.drop('price', axis = 1)
y = base.price
```

```python
X.head()
```

```python
type(X)
```

```python
y.head()
```

```python
type(y)
```

```python
previsores_colunas = colunas_usadas[1:17]
previsores_colunas
```

```python
import tensorflow as tf
```

```python
#Cria uma lista de colunas que serão usadas como previsores. a partir da lista de colunas usadas (que são strings)
colunas = [tf.feature_column.numeric_column(key = c) for c in previsores_colunas]
```

```python
print(colunas[10])
```

```python
#Separa a base de dados em treinamento e teste de forma que 70% dos dados serão usados para treinamento e 30% para teste
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.3)
```

```python
X_treinamento.shape
```

```python
y_treinamento.shape
```

```python
X_teste.shape
```

```python
#Prepara a estrutura que feedará o modelo com os dados de treinamento
#num_epochs = None é para ue o modelo treine até que o erro seja minimizado.
#shuffle = True é para que os dados sejam embaralhados a cada epoch
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x = X_treinamento, y = y_treinamento,
                                                        batch_size = 32, num_epochs = None, shuffle = True)
```

```python
#Prepara a estrutura que feedará o modelo com os dados de teste
#num_epochs = 10000 é para que o modelo treine 10000 vezes em cima dos dados de teste
#shuffle = False é para que os dados não sejam embaralhados
funcao_teste = tf.estimator.inputs.pandas_input_fn(x = X_teste, y = y_teste,
                                                   batch_size = 32, num_epochs = 10000, shuffle = False)
```

```python
#Cria o estimator que usa regressor linear e o prepara para lidar com as colunas que foram criadas
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
#O estimator é feedado com os dados de treinamento a partir da estrutura funcao_treinamento e realiza o treino internamente provavelmente usando SGE para minimizar o erro
#(Stochastic Gradient Descent)
regressor.train(input_fn=funcao_treinamento, steps = 10000)
```

```python
#O estimator é feedado com os dados de teste a partir da estrutura funcao_teste e é então avaliado a partir delas
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps = 10000)
```

```python
#O estimator é feedado com os dados de teste a partir da estrutura funcao_teste e é então avaliado a partir delas
#Aparentemente ele não guarda os dados internamente, pois posteriormente ele precisa ser realimentado para que então possa fazer previsões arbitrárias
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps = 10000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
#Prepara a estrutura que feedará o modelo com os dados de teste para realizar previsões arbitrarias
funcao_previsao = tf.estimator.inputs.pandas_input_fn(x = X_teste, shuffle = False)
```

```python
#Usa o estimator para realizar previsões a partir dos dados de teste que foram feedados a ele pela estrutura funcao_previsao e atribui o resultado a previsoes
#Nesse momento, provavelmente o estimator aprendeu os coeficientes da regressão linear e precisa receber os dados para realizar previsões em cima deles
#Já que ele não guarda os dados internamente dos outros passos
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
valores_previsoes = []

#Popula a lista valores_previsoes com as previsões realizadas pelo estimator a partir dos dados de teste feedados a ele pela estrutura funcao_previsao
for p in regressor.predict(input_fn=funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
import numpy as np

#Transforma a lista valores_previsoes em um array
#reshape(-1,1) é para que o array seja transformado em um array de uma coluna
#o -1 é para que o numpy entenda que o número de linhas é o número de elementos da lista
#o 1 é para que o numpy entenda que o número de colunas é 1
#Ou seja é como se "empilhase" tudo em uma única dimensão que é a coluna com todos os dados

#OBS: Não entendi para que isso é necessário, se valoes_previsoes já é um array já tem esse shape.
valores_previsoes = np.asarray(valores_previsoes).reshape(-1,1)
```

```python
#Aplica o desescalonamento nos valores de previsões
valores_previsoes = scaler_y.inverse_transform(valores_previsoes)
valores_previsoes
```
